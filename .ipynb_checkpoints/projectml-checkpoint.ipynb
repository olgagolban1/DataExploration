```python
import pandas as pd
import plotly.graph_objects as go

df_gender = pd.read_csv('dataexploproject/dataset/gifts_gender.csv')

df_gender_percent = df_gender.set_index('Gender').div(df_gender.set_index('Gender').sum(axis=1), axis=0) * 100

fig = go.Figure()

for column in df_gender_percent.columns:
    fig.add_trace(go.Bar(x=df_gender_percent.index, y=df_gender_percent[column], name=column))

fig.update_layout(
    barmode='stack',
    xaxis=dict(title='Gender'),
    yaxis=dict(title='Percentage'),
    title='Percentage of different gifts by gender'
)

fig.show()
```

```python
import pandas as pd
import plotly.subplots as sp
import plotly.graph_objects as go

df_age = pd.read_csv('datenset/archive/gifts_age.csv')

df_age = df_age.drop(columns=['SpendingCelebrating'])

fig = sp.make_subplots(rows=2, cols=3, subplot_titles=df_age['Age'].tolist(), specs=[[{'type': 'pie'}, {'type': 'pie'}, {'type': 'pie'}],
                                                                                   [{'type': 'pie'}, {'type': 'pie'}, {'type': 'pie'}]])

for i, age_group in enumerate(df_age['Age']):
    percentages = df_age[df_age['Age'] == age_group].squeeze()[1:]
    fig.add_trace(go.Pie(labels=percentages.index, values=percentages, hole=0.3, name=age_group),
                  row=(i // 3) + 1, col=(i % 3) + 1)

fig.update_layout(title_text='Percentage of gifts by age group', showlegend=False)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

df_historical = pd.read_csv('datenset/archive/historical_spending.csv')

fig = go.Figure()

categories = df_historical.columns[2:]
for category in categories:
    fig.add_trace(go.Scatter(x=df_historical['Year'], y=df_historical[category], mode='lines', name=category))

fig.add_trace(go.Scatter(x=df_historical['Year'], y=df_historical['PercentCelebrating'], mode='lines', name='PercentCelebrating', yaxis='y2'))

fig.update_layout(title_text='Historical course of spending behaviour (2010-2022)',
                  xaxis_title='Jahr', yaxis_title='Percentage/expenditure in US dollars',
                  yaxis2=dict(title='Percentage of those celebrating', overlaying='y', side='right', range=[0, 100]))

fig.show()
```

Korrelationsheatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Daten aus CSV-Datei lesen
file_path = '/Users/olgagolban/Desktop/uni/4. Semester/Data Exploration/dataexploproject/dataexploproject/dataset/gifts_gender.csv'
df = pd.read_csv(file_path)

# Setzen der Spalte 'Gender' als Index
df.set_index('Gender', inplace=True)

# Berechnung der Korrelationsmatrix
correlation_matrix = df.corr()

# Erstellen der Korrelationsheatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Korrelationsheatmap')
plt.show()
```

```python
import pandas as pd

# CSV-Daten einlesen
file_path = '/Users/olgagolban/Desktop/uni/4. Semester/Data Exploration/dataexploproject/dataexploproject/dataset/historical_spending.csv'  # Ersetzen Sie 'pfad_zu_ihrer_datei.csv' durch den tatsächlichen Dateipfad
df = pd.read_csv(file_path)

# Daten nach Jahren sortieren
df.sort_values(by='Year', inplace=True)

# Prozentuale Veränderung zum Vorjahr berechnen
df['PercentChange'] = df['PerPerson'].pct_change() * 100

# Transponieren des DataFrames und Spalten umsortieren
df_transposed = df.set_index('Year').T.sort_index(ascending=False)

# Tabelle anzeigen mit horizontalen Linien zwischen den Jahren
pd.set_option('display.width', 1000)  # Breite der Anzeige erhöhen, um Zeilen besser zu trennen
pd.set_option('display.max_rows', None)  # Anzeige aller Zeilen
pd.options.display.multi_sparse = False  # Multi-Index nicht überlappen lassen

display(df_transposed.style.set_table_styles([
    {'selector': 'tr:nth-of-type(odd)', 'props': 'background-color: #f4f4f4;'},
    {'selector': 'tr:nth-of-type(even)', 'props': 'background-color: #ffffff;'},
    {'selector': 'thead th', 'props': 'background-color: #3498db; color: #ffffff;'},
    {'selector': 'th, td', 'props': 'border-bottom: 1px solid #ddd;'},
]))


```
